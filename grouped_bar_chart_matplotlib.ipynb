```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
cost_objects = ["Cement", "Steel", "Bricks", "Project Manager", "Plumber", "Supervisor",
                "Painting", "Security", "Electrical", "Crane", "Fork-lift", "Genie boom",
                "Hotel", "Travel", "Insurance"]

variance = [-20, -10, -5, -18, -8, -6, -25, -15, -12, -18, -10, -9, -8, -8, -6]

cost_types = ["Material", "Material", "Material", "Labor", "Labor", "Labor",
              "Sub-contracting", "Sub-contracting", "Sub-contracting",
              "Equipment", "Equipment", "Equipment", "Sundry", "Sundry", "Sundry"]

colors = {"Material": "blue", "Labor": "green", "Sub-contracting": "red",
          "Equipment": "purple", "Sundry": "orange"}

# Assign colors based on cost types
bar_colors = [colors[ct] for ct in cost_types]

# Plot Horizontal Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.barh(cost_objects, variance, color=bar_colors)

plt.xlabel("Variance at Completion (%)")
plt.ylabel("Cost Objects")
plt.title("Forecasted Cost Overruns by Cost Object")

# Reverse y-axis to keep the highest variance on top
plt.gca().invert_yaxis()

# Legend
import matplotlib.patches as mpatches
legend_patches = [mpatches.Patch(color=colors[key], label=key) for key in colors]
plt.legend(handles=legend_patches, title="Cost Type")

plt.show()
```
