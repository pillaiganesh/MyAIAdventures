```python
import matplotlib.pyplot as plt

def create_pie_chart(data):
    categories = [defect["category"] for defect in data["defects"]]
    counts = [defect["count"] for defect in data["defects"]]

    plt.pie(counts, labels=categories, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title("Defects Distribution")
    plt.show()

# Your JSON data
defects_data = {
    "defects": [
        {"category": "Design Defect", "count": 10},
        {"category": "Code Bug", "count": 100},
        {"category": "User Misunderstanding", "count": 50}
    ]
}

# Create and display the pie chart
create_pie_chart(defects_data)
```
